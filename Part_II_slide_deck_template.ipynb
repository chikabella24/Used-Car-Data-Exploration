## Dataset Overview
The dataset is from https://www.kaggle.com. It is  about an automobile company XYZ from India which aspires to enter the Indian used-car market by setting up their company locally to give competition to their counterparts. The Used Car Data consists of 7906 observations and 18 features('Sales_ID','name','year','selling_price','km_driven','Region','State or Province',''City','fuel',seller_type','transmission','owner','mileage','engine','max_power','torque','seats','sold'.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
used_car =pd.read_csv('UserCarData.csv')
```

## The Relationship between Regions and Price

This describes how the Regions (East,Central,West,South) is significant in predicting the price of a used car
It shows that the east region had the highest selling price for cars here we can see the presence of an outlier in the east region that increases the price to the highest level,the  followed by the south region which also have an outlier


```python
#creating a plot to show the relationship of Region and selling price
plt.figure(figsize = [15,5])
plt.subplot(1,2,1)
ax = plt.gca() 
ax.ticklabel_format(useOffset=False, style='plain')
sns.boxplot(data = used_car,x = 'Region',y = 'selling_price' );
ax.set_title('Region by Selling Price')
ax.set_xlabel('Regions')             
ax.set_ylabel('Selling Price ($)')
plt.subplot(1,2,2)
ax = plt.gca() 
ax.ticklabel_format(useOffset=False, style='plain')
sns.violinplot(data = used_car,x = 'Region',y = 'selling_price' );
ax.set_title('Region by Selling Price')
ax.set_xlabel('Regions')             
ax.set_ylabel('Selling Price ($)');
```

## The Relationship between Name and Price
This describes which brand names are a good predictor of price
It indicates that the name Mariuti,Hyundai,Bmw  nd Mahindra command a higher selling price  and so its a good predictor of price


```python
#calculate sum of selling Price for each Brand name of used car
name_by_selling_price =used_car.groupby('name').selling_price.sum().sort_values(ascending= False)
```

```python
#plot to display the 
plt.figure(figsize=(20,15))
#create bar plot with custom aesthetics
base_color = sns.color_palette()[2]
ax = plt.gca()
ax.ticklabel_format(useOffset=False, style='plain', axis='y')
name_by_selling_price.plot(kind='bar', title='Brand Names by Selling Price',color = base_color,
               ylabel='Selling Price ($)', xlabel='Brand Names')

#rotate x-axis ticks vertically
plt.xticks(rotation=45);
```

## The Relationship between Owner and Price

This describes the relationhip between the owner and the price

it indicates the the owner on the first_owner has a high significant relationship between selling price of the car so its a good predictor of price. Most of the cars are sold are sold on first_owner basis

```python
#plotting the relationship between the owner and selling price
plt.figure(figsize = [20, 5]) 
base_color=sns.color_palette()[0]
ax = plt.gca()
ax.ticklabel_format(useOffset=False, style='plain')
owner_type = used_car.groupby('owner').selling_price.sum()
sns.barplot(owner_type,owner_type.index,color=base_color)
ax.set_title('Owner by Selling Price')
ax.set_xlabel('Selling Price ($)')             
ax.set_ylabel('Owner Types');
```

## The Relationship between Transmission and Price
This describes the relationship between transmission and price

It indicates that used car Transmission mode Manual are being sold more with a  higher price

```python
#calculate sum of selling Price for each transmission type of used car
transmission_by_sold = used_car.groupby('transmission').selling_price.sum().sort_values(ascending = False)
```

```python
#plotting a plot to show the mode of transmission and the selling  price
plt.figure(figsize=(12,6))
#create bar plot with custom aesthetics
base_color = sns.color_palette()[2]
ax = plt.gca()
ax.ticklabel_format(useOffset=False, style='plain', axis='y')
transmission_by_sold.plot(kind='bar', title='Mode of Transmission by Selling Price',color = base_color,
               ylabel='Selling Price (Rs)', xlabel='Transmission Modes');
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='white', size=10)
plt.show()
```

## The Relationship between km_driven and Price
This indicates a strongly negative  correlation, the distance driven on the used car does not predict the pricing of the car.

As the distance driven increases the selling price doesnt, the selling price is still around 500000. Except in a few case where due to presence of an outlier we do have high price.

Most of the cars have not been driven for long

```python
# plot to show the relationship betwwen distsace driven and selling price
plt.figure(figsize = (12,6))
ax = plt.gca()
ax.ticklabel_format(useOffset=False, style='plain')
sns.scatterplot(used_car.engine, used_car.km_driven)
ax.set_title('Distance Driven by Selling Price')
ax.set_xlabel('Distance Driven (km)')             
ax.set_ylabel('Selling Price (Rs)');
```

## The relationship between the fuel types and the selling price
This describes how the fuel types (Diesel,Petrol,CNG,LPG) is significant in predicting the price of a used car

THis shows tha the Diesel is a predictor of price of the used car as it commands the highest selling price

and we went further to show the proprtion of its being sold and unsold
 

```python
plt.figure(figsize =(12,8))
ax = plt.gca()
ax.ticklabel_format(useOffset=False, style='plain')
ax.set_title('Fuel Types by Selling Price')
ax.set_ylabel('Selling Price')             
ax.set_xlabel('Fuel Types');
sns.barplot(data = used_car,x = 'fuel',y = 'selling_price',hue ='sold',palette = 'plasma');
```
